```python
import vk_api
import pandas as pd
import datetime
from tqdm import tqdm
import numpy as np
```

```python
user_info = pd.read_csv('data/vk/user_info.csv')

wall_comments = pd.read_csv('data/vk/wall-comments.csv')

wall_likes_and_reposts = pd.read_csv('data/vk/wall-likes-and-reposts.csv')
wall_likes_and_reposts.drop('Unnamed: 0', axis = 1, inplace = True)

wall = pd.read_csv('data/vk/wall.csv')
```

```python
wall_comments.head()
```

```python
wall_comments_gr = wall_comments.groupby(['user_id', 'post_id'], as_index = False)['likes'].mean()[['user_id', 'post_id']]
wall_comments_gr['comment'] = 1
wall_comments_gr.head()
```

```python
wall_likes_and_reposts['like'] = 1

wall_likes_and_reposts.head()
```

```python
coms_reps_likes = pd.merge(
    left = wall_comments_gr,
    right = wall_likes_and_reposts,
    how = 'outer',
    on = ['user_id', 'post_id']
)

coms_reps_likes.head()
```

```python
import itertools

post_id = list(set(wall['post_id']))
user_id = list(set(user_info['user_id']))

all_pairs = []

for i in itertools.product(user_id, post_id):
    all_pairs.append(i)
    
all_user_and_posts = pd.DataFrame(all_pairs)
all_user_and_posts.columns = ['user_id', 'post_id']
all_user_and_posts.head()
```

```python
total = pd.merge(
    left = all_user_and_posts,
    right = user_info,
    how = 'outer',
    on = 'user_id'
)

try:
    total.drop('Unnamed: 0', axis = 1, inplace = True)
except Exception:
    pass

total = pd.merge(
    left = total,
    right = coms_reps_likes,
    how = 'outer',
    on = ['user_id', 'post_id']
)

total['comment'].fillna(0, inplace = True)
total['is_repost'].fillna(0, inplace = True)
total['like'].fillna(0, inplace = True)

total['action'] = total[['comment','is_repost','like']].apply(np.max, axis = 1)

total.head()
```

```python
post_id_commens_count = total.groupby(['post_id'], as_index = False)['comment'].sum()
post_id_is_repost_count = total.groupby(['post_id'], as_index = False)['is_repost'].sum()
post_id_like_count = total.groupby(['post_id'], as_index = False)['like'].sum()
post_id_action_count = total.groupby(['post_id'], as_index = False)['action'].sum()

for i in (post_id_is_repost_count, post_id_like_count, post_id_action_count):
    post_id_commens_count = post_id_commens_count.merge(i, how = 'outer', on = 'post_id')

post_id_commens_count.columns = ['post_id', 'post_comm_count', 'post_repost_count', 'post_like_count','post_action_count']
post_id_commens_count.head()
```

```python
final = total.merge(post_id_commens_count, how = 'outer', on = 'post_id')

final.drop('city', axis = 1, inplace = True)
final.drop('relation', axis = 1, inplace = True)
```

```python
final.head()
```

```python
final.loc[~final['country'].isin(['Россия', 'Украина', 'Казахстан']), 'country'] = None
```

```python
final = pd.get_dummies(final, columns = ['country'])
final = pd.get_dummies(final, columns = ['sex'])
```

```python
final = final[['user_id', 'post_id', 'sex_1.0', 'sex_2.0', 'years', 'country_Казахстан', 'country_Россия', 'country_Украина', 
              'post_comm_count','post_repost_count','post_like_count','post_action_count', 'comment','is_repost','like','action']]
final.head()
```
