```python
import vk_api
import pandas as pd
import datetime
from tqdm import tqdm
```

```python
vk_session = vk_api.VkApi(token='')
```

```python
vk = vk_session.get_api()
```

```python
# group_members список пользователей, состоящих в группе:
tools = vk_api.VkTools(vk_session)
group_members_id = tools.get_all('groups.getMembers', 1000, {'group_id': 108367462})
group_members_id = group_members_id['items']
print(len(group_members_id))
```

```python
# Информация о пользователях:

all_dict_of_users = []
for user_id in tqdm(group_members_id):

    
    try:
        dict_of_users = vk.users.get(user_id=user_id, order='hints', 
                             fields=['sex, bdate, city, country, education, universities, schools, relation'])
        all_dict_of_users.extend(dict_of_users)
    except Exception:
        continue

list_for_df = []
for k in all_dict_of_users:
    list_ = []

    try:
        list_.append(k['id'])
    except Exception:
        list_.append(None)
    try:
        list_.append(k['sex'])
    except Exception:
        list_.append(None)
    try:
        list_.append(k['bdate'])
    except Exception:
        list_.append(None)
    try:
        list_.append(k['relation'])
    except Exception:
        list_.append(None)
    try:
        list_.append(k['city']['title'])
    except Exception:
        list_.append(None)
    try:
        list_.append(k['country']['title'])
    except Exception:
        list_.append(None)
    try:
        list_.append(k['schools'][0]['year_graduated'])
    except Exception:
        list_.append(None)
    try:
        list_.append(k['universities'][0]['chair_name'])
    except Exception:
        list_.append(None)
    try:
        list_.append(k['universities'][0]['faculty_name'])
    except Exception:
        list_.append(None)
    try:
        list_.append(k['universities'][0]['graduation'])
    except Exception:
        list_.append(None)
    list_for_df.append(list_)

user_info = pd.DataFrame(list_for_df)
user_info.columns = ['user_id', 'sex', 'birthday', 'relation', 'city', 'country',
                                'school_year_graduated', 'universities_chair_name', 
                                'universities_faculty_name', 'universities_graduation']
```

```python
user_info['today'] = '11.12.2016'
```

```python
user_info
```

```python
my_df = user_info.copy()

for i, row in my_df[my_df['birthday'].notnull()].iterrows():
    if len(row['birthday']) <= 5:
        my_df.loc[i, 'birthday'] = None
        
my_df['birthday'] = pd.to_datetime(my_df['birthday'], format = '%d.%m.%Y')
my_df['today'] = pd.to_datetime(my_df['today'], format = '%d.%m.%Y')
my_df['years'] = my_df['today'] - my_df['birthday']

def get_years(x):
    try: 
        return x.days/364.25
    except:
        return x
    
my_df['years'] = my_df['years'].apply(get_years)


for i, row in my_df[my_df['birthday'].notnull() == False].iterrows():
    if (row['birthday'] is None) & (row['school_year_graduated'] is not None):
        my_df.loc[i, 'birthday'] = 17 - (my_df.loc[i, 'school_year_graduated'] - 2016)
        
for i, row in my_df[my_df['birthday'].notnull() == False].iterrows():
    if (row['birthday'] is None) & (row['universities_graduation'] is not None):
        my_df.loc[i, 'birthday'] = 22 - (my_df.loc[i, 'universities_graduation'] - 2016)
        
```

```python
my_df = my_df[['user_id', 'sex', 'years', 'relation', 'city', 'country']]
my_df
```

```python
my_df.to_csv('data/vk/user_info.csv')
```
