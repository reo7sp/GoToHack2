```python
mit_df = pd.read_csv('data/vk/mit_table.csv')
mit_df.head()
```

```python
import numpy as np
```

```python
from tqdm import tqdm
```

```python
X_predf = []

for user_id, df in tqdm(info_df.groupby('user_id')):
    f = df.iloc[0]
    row = [user_id, f['sex'], f['years'], f['relation'], f['country_Казахстан'], f['country_Россия'], f['country_Украина']]
    inst_a = np.array([0.0] * 15)
    inst_b = np.array([0.0] * 15)
    size = 0
    for i, r in df.iterrows():
        if r.action == 0:
            continue
        
        # groups
        try:
            groups = mit_df[mit_df.user_id == user_id].group_id.values
            inst_a += descs_lda_groups_df.loc[user_id]
        except:
            pass
        
        # posts
        post_id = r['post_id']
        ind = -1
        for j, w in wall.iterrows():
            if w.post_id == post_id:
                ind = j
                break
        inst_b += lda_posts_df.iloc[ind]
        
        size += 1
        
    inst_a /= size
    inst_b /= size
    
    row.extend(inst_a)
    row.extend(inst_b)
    
    X_predf.append(row)
```
