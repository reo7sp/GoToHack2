```python
import vk_api
import pandas as pd
```

```python
vk_session = vk_api.VkApi(token='')
```

```python
vk = vk_session.get_api()
```

```python
# group_members список пользователей, состоящих в группе:
response = vk.groups.getMembers(group_id=108367462)
group_members_id = response['items']
```

```python
# Друзья каждого пользовтеля, состоящего в группе:

all_dict_of_friends = []
for user_id in group_members_id:
    #if vk.friends.get(user_id=user_id)['count'] < 500:
    try:
        dict_of_friends = vk.friends.get(user_id=user_id, order='hints', 
                             fields=['sex, bdate, city, country, education, universities, schools, relation'])
    
    dict_of_friends['user_id'] = user_id
    all_dict_of_friends.append(dict_of_friends)
    list_for_df = []
    for k in all_dict_of_friends:
        for i in range(len(k['items'])):
            list_ = []
            
            try:
                list_.append(k['user_id'])
            except Exception:
                list_.append(None)
            try:
                list_.append(k['items'][i]['id'])
            except Exception:
                list_.append(None)
            try:
                list_.append(k['items'][i]['sex'])
            except Exception:
                list_.append(None)
            try:
                list_.append(k['items'][i]['bdate'])
            except Exception:
                list_.append(None)
            try:
                list_.append(k['items'][i]['relation'])
            except Exception:
                list_.append(None)
            try:
                list_.append(k['items'][i]['city']['title'])
            except Exception:
                list_.append(None)
            try:
                list_.append(k['items'][i]['country']['title'])
            except Exception:
                list_.append(None)
            try:
                list_.append(k['items'][i]['schools'][0]['year_graduated'])
            except Exception:
                list_.append(None)
            try:
                list_.append(k['items'][i]['universities'][0]['chair_name'])
            except Exception:
                list_.append(None)
            try:
                list_.append(k['items'][i]['universities'][0]['faculty_name'])
            except Exception:
                list_.append(None)
            try:
                list_.append(k['items'][i]['universities'][0]['graduation'])
            except Exception:
                list_.append(None)
            list_for_df.append(list_)

user_and_frends_info = pd.DataFrame(list_for_df)
user_and_frends_info.columns = ['user_id', 'friend_id', 'sex', 'birthday', 'relation', 'city', 'country',
                                'school_year_graduated', 'universities_chair_name', 
                                'universities_faculty_name', 'universities_graduation']
```
